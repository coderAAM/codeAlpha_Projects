```python
# Install kagglehub (only needed in Colab)
!pip install kagglehub

# Import libraries
import pandas as pd
import os
import kagglehub

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Download dataset
path = kagglehub.dataset_download("saurabh00007/iriscsv")
print("Dataset path:", path)

# Check files
print("Files:", os.listdir(path))

# Load dataset (name check karo)
file_path = os.path.join(path, "Iris.csv")
data = pd.read_csv(file_path)

print(data.head())

# Features & labels
X = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = data['Species']

# Encode labels
le = LabelEncoder()
y = le.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = KNeighborsClassifier(n_neighbors=3)

# Train
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Custom test
sample = [[5.1, 3.5, 1.4, 0.2]]
# Convert the sample to a DataFrame with the same column names as X_train
sample_df = pd.DataFrame(sample, columns=X_train.columns)
prediction = model.predict(sample_df)

print("Predicted Flower:", le.inverse_transform(prediction))
```
